```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Create an empty list to store the details of the topics
topic_details = []

# Prompt the user to enter a topic and convert it to lowercase
topic = input('Enter topic: ').lower()

# Prompt the user to enter the number of pages to scrape
page_number = int(input('Enter how many pages you want to scrape: '))

# Display the completion status of scraping each page
print('Web Scraping Page Number completion status:')

# Iterate over the range of page numbers
for i in range(1, page_number + 1):
    # Create the URL by replacing the topic and page number in the template URL
    url = 'https://www.bbc.co.uk/search?q={}&d=SEARCH_PS&page={}'.format(topic, i)
    
    # Send a GET request to the URL and get the response
    response = requests.get(url)
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the container element that holds the details of each article
    #details_elements = soup.find_all(class_='ssrcss-1020bd1-Stack e1y4nx260')
    details_elements = soup.find_all('div', class_='ssrcss-53phst-Promo ett16tt0')
    #print(details_elements)
    
    # If no details elements are found, break the loop
    if details_elements is None:
        break
    else:
        # Iterate over each detail element
        for element in details_elements:
            # Extract the necessary information from each element
            Title = element.find('p', class_='ssrcss-6arcww-PromoHeadline e1f5wbog5').text.strip()
            Description = element.find('p',class_='ssrcss-1q0x1qg-Paragraph eq5iqo00').text.strip()
            #Programmes = element.find('span',class_='ssrcss-1if1g9v-MetadataText ecn1o5v1').text.strip()
            Programmes = []
            Programmes_elements = element.find_all('span', class_='ssrcss-1if1g9v-MetadataText ecn1o5v1')
            for Programmes_element in Programmes_elements:
                Programmes.append(Programmes_element.text.strip() if Programmes_element else "")
            Programmes_str = '. '.join(Programmes)
            href = element.find('a', class_='ssrcss-rl2iw9-PromoLink e1f5wbog1')['href']
            
            topic_details.append({
                'Title': Title,
                'Description': Description,
                'Programmes': Programmes,
                'Refrence_URL': href
            })
            
        # Display the completion status of scraping the current page
        print(i, end=',')     
```

```python
## Create a DataFrame from the team details
df = pd.DataFrame(topic_details)

# Save the DataFrame as CSV
df.to_csv('{}_details_bbc_{}.csv'.format(topic,page_number), index=False)
print('{}_details_bbc_{}.csv'.format(topic,page_number)+' is saved')
print()
print()
print()
# Print the DataFrame
df
```
